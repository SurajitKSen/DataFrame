```python
import os
import zipfile

def extract_all_zips(folder_path, extract_to):
    # Ensure the extraction directory exists
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)

    # Loop through all files in the given directory
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.zip'):
            file_path = os.path.join(folder_path, file_name)
            
            # Extract the zip file
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(extract_to)
                print(f'Extracted {file_name} to {extract_to}')

# Example usage
folder_path = 'D:\OneDrive_2024-02-13\Python\DataFrame'
extract_to = 'D:\OneDrive_2024-02-13\Python\DataFrame'

extract_all_zips(folder_path, extract_to)
```
