```python
from pptx import Presentation
from pptx.util import Pt

def change_font_to_default(pptx_file, output_file):
    # Load the presentation
    prs = Presentation(pptx_file)

    # Define the default font name
    default_font_name = 'Arial'  # You can change this to your preferred default font

    for slide in prs.slides:
        for shape in slide.shapes:
            if shape.has_text_frame:
                for paragraph in shape.text_frame.paragraphs:
                    for run in paragraph.runs:
                        run.font.name = default_font_name

    # Save the modified presentation
    prs.save(output_file)

# Usage
input_pptx_file = 'DataFrames and Joins Final2.pptx'
output_pptx_file = 'DataFrames and Joins.pptx'
change_font_to_default(input_pptx_file, output_pptx_file)
```
