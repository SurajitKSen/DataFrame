# Python Pandas DataFrame

# Defination:
Pandas DataFrame is a widely used data structure which works with a two-dimensional array with labeled axes (rows and columns). 
DataFrame is defined as a standard way to store data that has two different indexes, i.e., row index and column index. 

# This means a dataframe is a two-dimensional labelled data structure like a table in database.

# Properties:
The columns can be heterogeneous types like int, bool, and so on.
It can be seen as a dictionary of Series structure where both the rows and columns are indexed. It is denoted as "columns" in case of columns and "index" in case of rows.

#### > Each column can have a different types of values.
#### > This dataframe falls under pandas library. (pandas is an open-source library)
#### > Heterogeneous data type.

# How to create a DataFrame

## Create an empty dataframe

```python
import pandas as pd
dataframe = pd.DataFrame()

#view DataFrame
print(dataframe)
```

# Ways to cerate a dataframe

We can create a DataFrame using following ways:

. Lists (Should be of same length)

. Dict of List (Should be of same length)

. Dict of Series (Can be of any length)

. Another DataFrame

. From External Files like CSV, JSON, Excel Sheet

## Create a dataframe using List

```python
import pandas as pd
# list = ['','']
list = ['Python', 'Pandas']

# Calling DataFrame constructor on list  
DataFrame1 = pd.DataFrame(list)
#view DataFrame
print(DataFrame1)

#view type of DataFrame
print(type(DataFrame1))


print("\n")

# Another way List of List
import pandas as pd
data = [['Alex',10],['Bob',12],['Clarke',13]]
DataFrame2 = pd.DataFrame(data,columns=['Name','Age'])
#view DataFrame
print(DataFrame2)
```

## Create a DataFrame from Dict of List

```python
import pandas as pd
# dict = {'':[],'':[]} or {key:'value'}
dict = {'ID' :[101, 102, 103], 'Department' :['B.Sc','B.Tech','M.Tech',]}

# Calling DataFrame constructor on dictionary
DataFrame3 = pd.DataFrame(dict)
print (DataFrame3)
```

# What is a Series ?

```python
import pandas as pd  
# A series is one-dimensional array that can hold data of any type.
a = [1,2,3]
var = pd.Series(a)
print(var)

print("\n")

# How to return a single value from the Series
print(var[0])
print(var[1])

print("\n")

# How to change the index? By using index argument 
list_info = [1, 2, 3]
info = pd.Series(list_info, index = ["x", "y", "z"])
print(info)
```

## Create a DataFrame from Dict of Series

```python
# How to return a selected column from the DataFrame?
info = {'one' : pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f']),  
   'two' : pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])}  
  
DataFrame4 = pd.DataFrame(info)  
print (DataFrame4)
```

## Create a DataFrame with a list of dictionaries, row indices, and column indices

```python
import pandas as pd
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20}]

# With two column indices, values same as dictionary keys
DataFrame5 = pd.DataFrame(data, index=['first', 'second'], columns=['a', 'b'])

#With two column indices with one index with other name
DataFrame6 = pd.DataFrame(data, index=['first', 'second'], columns=['a', 'b1'])
print(DataFrame5)
print("\n")
print(DataFrame6)
```

# Create a DataFrame from existing DataFrame

```python
old_df = pd.DataFrame({'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
                       'points': [18, 22, 19, 14, 14, 11, 20, 28],
                       'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                       'rebounds': [11, 8, 10, 6, 6, 7, 9, 12]})

#view DataFrame
print(old_df)

print("\n")

#create new DataFrame from existing DataFrame
new_df = old_df[['points','rebounds']].copy()

#view new DataFrame
print(new_df)
print("\n")
print(type(new_df))


#create new DataFrame from existing DataFrame but do not include 1 column
new_df2 = old_df.drop('points', axis=1)    # axis=0 (default): Refers to rows. If you want to drop rows from the DataFrame, you use this setting.
                                           # axis=1: Refers to columns. If you want to drop columns from the DataFrame, you use this setting.

print("\n")
#view new DataFrame
print(new_df2)
print("\n")
print(type(new_df2))


#create new DataFrame from existing DataFrame but transpose
print("\n")
transposed_df = new_df2.transpose()
#view new DataFrame
print(transposed_df)
print(type(transposed_df))
```

```python
#create new DataFrame from existing DataFrame

# Original DataFrame
old_df2 = pd.DataFrame({ 'team': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'], 'points': [18, 22, 19, 14, 14, 11, 20, 28]})

# DataFrame to append
new_data2 = pd.DataFrame({ 'team': ['C', 'C'], 'points': [25, 30] })

new_df = pd.concat([old_df2, new_data2], ignore_index=True)
# Display the new DataFrame
print(new_df)
```

#### Select specific columns from the dataset
Also specify that the decimal is a comma (European style) and not a dot (the norm outside Europe)

```python
required_cols = ['Country', 
                 'Population', 
                 'Area (sq. mi.)', 
                 'Birthrate', 
                 'Deathrate']
```

```python
countries = pd.read_csv('countries_of_the_world.csv', 
                        usecols = required_cols, 
                        decimal = ',')
```

```python
countries.head()
```

```python
countries['Country']
```

```python
countries['Country'].head()
```

```python
countries['Country'][206]
```

#### Slicing notation applies to Dataframes too
Pandas is built on top of NumPy so some operations overlap. The Dataframe is effectively a 2-D array with the column being the first dimension and the row is the second

```python
countries['Country'][105:112]
```

#### Select range of rows from specific columns
Note the array of string column names passed as opposed to the range of integer row indices

```python
countries[['Country', 'Birthrate']][105:112]
```

### Adding own column with convertion

```python
# is converting the area of countries from square miles to square kilometers

countries['Area'] = countries['Area (sq. mi.)'] * 1.609 * 1.609
countries.head()
```
