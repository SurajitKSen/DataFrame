```python
import json
import os
import random
import string
import uuid
from datetime import datetime, timedelta
import hashlib

# generate_users_json.py

FIRST_NAMES = ["Alex", "Jamie", "Taylor", "Jordan", "Casey", "Riley", "Morgan", "Sam", "Dana", "Lee"]
LAST_NAMES = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis", "Martinez", "Clark"]
DOMAINS = ["example.com", "mail.com", "test.org", "demo.net"]
STREETS = ["Main St", "Oak Ave", "Pine Rd", "Maple Dr", "Cedar Ln"]
CITIES = ["Springfield", "Rivertown", "Greenville", "Lakeside", "Hillview"]
ROLES = ["user", "admin", "editor", "viewer"]
STATUSES = ["active", "inactive", "banned", "pending"]

def random_name():
    return f"{random.choice(FIRST_NAMES)} {random.choice(LAST_NAMES)}"

def random_email(name):
    local = "".join(ch for ch in name.lower() if ch.isalpha())
    return f"{local}.{random.randint(1,999)}@{random.choice(DOMAINS)}"

def random_phone():
    return f"+1-{random.randint(200,999)}-{random.randint(200,999)}-{random.randint(1000,9999)}"

def random_address():
    return {
        "street": f"{random.randint(1,9999)} {random.choice(STREETS)}",
        "city": random.choice(CITIES),
        "zip": f"{random.randint(10000,99999)}",
        "country": "USA"
    }

def random_date_between(start_year=1950, end_year=2005):
    start = datetime(start_year, 1, 1)
    end = datetime(end_year, 12, 31)
    delta = end - start
    return (start + timedelta(days=random.randint(0, delta.days))).date().isoformat()

def hash_password(raw):
    return hashlib.sha256(raw.encode("utf-8")).hexdigest()

def make_user():
    name = random_name()
    raw_password = "".join(random.choices(string.ascii_letters + string.digits, k=12))
    return {
        "id": str(uuid.uuid4()),
        "name": name,
        "email": random_email(name),
        "password_hash": hash_password(raw_password),
        "phone": random_phone(),
        "dob": random_date_between(),
        "registered_at": (datetime.utcnow() - timedelta(days=random.randint(0, 3650))).isoformat() + "Z",
        "address": random_address(),
        "roles": random.sample(ROLES, k=random.randint(1,2)),
        "status": random.choice(STATUSES),
        "preferences": {
            "newsletter": random.choice([True, False]),
            "sms_alerts": random.choice([True, False]),
            "theme": random.choice(["light", "dark"])
        },
        "metadata": {
            "score": round(random.uniform(0, 100), 2),
            "tags": random.sample(["beta", "vip", "trial", "partner", "internal"], k=random.randint(0,2))
        }
    }

def generate_users(count=10, filepath="users.json", seed=None):
    if seed is not None:
        random.seed(seed)
    users = [make_user() for _ in range(count)]
    dirname = os.path.dirname(filepath)
    if dirname:
        os.makedirs(dirname, exist_ok=True)
    with open(filepath, "w", encoding="utf-8") as f:
        json.dump({"generated_at": datetime.utcnow().isoformat() + "Z", "count": len(users), "users": users}, f, indent=2)
    return filepath

def read_users(filepath="users.json"):
    with open(filepath, "r", encoding="utf-8") as f:
        return json.load(f)
```

```python
if __name__ == "__main__":
    # Always place file in current working directory 
    current_dir = os.getcwd() 
    file_path = os.path.join(current_dir, "users.json") 
    # Generate and save users 
    out = generate_users(count=25, filepath=file_path, seed=42) 
    data = read_users(out) 
    print(f"Saved {data.get('count')} users to {out}") 
    print("Sample user:") 
    print(json.dumps(data["users"][0], indent=2))
```

```python
import json
import os

def read_users_from_cwd(filename="users.json"):
    # Build full path to file in current working directory
    current_dir = os.getcwd()
    file_path = os.path.join(current_dir, filename)

    # Read JSON data
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    return data, file_path

if __name__ == "__main__":
    users_data, path = read_users_from_cwd()
    print(f"Loaded {users_data['count']} users from {path}")
    print("First user entry:")
    print(json.dumps(users_data["users"][0], indent=2))
```
