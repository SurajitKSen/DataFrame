```python
import pandas as pd
```

```python
# Reading an Excel Sheet
```

```python
df = pd.read_csv('Employee.csv')
```

```python
# Exploring Dataframe Attributes
```

```python
print(df.shape)
print(df.columns)
# print(countries.info)
```

```python
# Select/Filter specific columns from the dataset
```

```python
selected_columns = df[['Country', 'Salary']]
selected_columns
```

```python
afghanistan_df = df[df['Country'] == 'Afghanistan']
afghanistan_df
```

```python
Modified_employee_id_df = df[df['EmployeeID']!= 3]
Modified_employee_id_df
```

```python
# Filtering rows based on a condition
filtered_df = df[df['Salary'] > 5000]
filtered_df
```

```python
# Sorting
sorted_df = df.sort_values(by='EmployeeID', ascending=True)
sorted_df
```

```python
# Aggregation
aggregated_df = df.groupby('Salary').sum()
new_aggregated_df = pd.DataFrame(aggregated_df)

new_aggregated_df
```

```python
# Merging DataFrames: Concat, Merge Functions

# Sample DataFrame 1
df1 = df[['EmployeeID', 'Name', 'Country']]

# Sample DataFrame 2
df2 = df[['EmployeeID', 'Salary']]

# Concat
concat_df = pd.concat([df1, df2], axis=1)

# Merge
merged_by_left_join_df = pd.merge(df1, df2, on='EmployeeID', how = 'left')
merged_by_right_join_df = pd.merge(df1, Modified_employee_id_df, on='EmployeeID', how = 'right') # inner, outer


print(concat_df.head())
```

```python
print(merged_by_left_join_df.head())
```

```python
print(merged_by_right_join_df.head())
```

```python
# Handling Missing Data: Identify Missing Values, Fill Missing Values, Drop Missing Values

# Identifying missing values
missing_values = df.isnull().sum()

# Fill missing values
df_filled = df.fillna(0)

# Drop missing values
df_dropped = df.dropna()

print('Missing Values:')
print(missing_values)
print('DataFrame with Filled Missing Values:')
print(df_filled.head())
print('DataFrame with Dropped Missing Values:')
print(df_dropped.head())
```

```python
# Exporting DataFrame to Excel ".CSV"

df_dropped.to_csv('Employee_output.csv', index=False)
```
